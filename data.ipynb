```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data ={ 
    "salary":[25000,35000,55000,67000,90000,12300]
       }
df=pd.DataFrame(data)
df.head()
df.shape
```

```python
plt.plot(df["salary"], color= "red", marker = "o")
plt.grid()
plt.show()
```

```python
plt.hist(df["salary"], bins= 5, color="green")
plt.show()
```

```python
plt.boxplot(df["salary"])
plt.show()
```
